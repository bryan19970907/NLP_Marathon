#### Question:
若你是決策樹，下列兩種分類狀況(a,b)，你會選擇哪種做分類？為什麼？

### Answer:
(a)

I(left) = -( 3/4*log(3/4) + 1/4*log(1/4) ) = 0.81

I(right) = -(1/4*log(1/4) + 3/4*log(3/4) ) = 0.81

IG = 0.19

(b)

I(left) = -(3/7*log(3/7) + 4/7*log(4/7) ) = 0.523 + 0.46 = 0.98

I(right) = 0

IG = 1 - 0.875*0.98 = 1 - 0.8575 = 0.1425

選(a)，因為information gain最大(或亂度下降最多)
